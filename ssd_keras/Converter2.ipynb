```python
import pandas as pd

df = pd.read_csv("../object-detection-crowdai/labels.csv")
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
df.head()
```

```python
le.fit(df["Label"])
le.classes_
```

```python
labels = le.transform(df["Label"]) + 1
```

```python
with open("../object-detection-crowdai/train_labels.csv", "w") as f:
    for i,l in enumerate(labels):
          
        f.write("%s,%d,%d,%d,%d,%d\n" %(df['Frame'][i],df['xmin'][i]/4,df['ymin'][i]/4,df['xmax'][i]/4,df['ymax'][i]/4,l))
    
```

```python
import glob 
filelist = glob.glob("../object-detection-crowdai/*.jpg")
import cv2
for filename in filelist:
    img = cv2.imread(filename)

    img = cv2.resize(img, (480,300))

    cv2.imwrite(filename, img)
```
